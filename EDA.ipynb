**Handling missing value**

```python
#Handling missing value
#Nan replaced by mean in Calories

df['Calories'] = df['Calories'].fillna(df['Calories'].mean())
df.head(25)

```

```python
#Nan replaced by methods rather than mean in Calories
df['Calories'] = df['Calories'].fillna(method='pad')
df.head(25)
```

```python
##Nan replaced by median & mode rather than mean in Calories
df['Calories'] = df['Calories'].fillna(df['Calories'].median())
df.head(25)

# mode
df['Calories'] = df['Calories'].fillna(df['Calories'].mode())
df.head(25)

```
