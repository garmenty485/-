```python
score = 0
incorrect_answers = []

# Iterating user's inputs　(考試作答開始)
for i in range(len(correct_answers)):
    user_answer = input("Enter your answer for question {} (A/B/C/D): ".format(i+1))
    if user_answer == correct_answers[i]:
        score += 1
    else:
        incorrect_answers.append(i+1)

# Checking answers
print("Score:", score)
if incorrect_answers:
    print("Incorrect answers:", incorrect_answers)
else:
    print("Congratulations! You got all the answers correct.")
```
